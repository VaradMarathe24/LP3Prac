```python
# Recursive Fibonacci function (for a single value)
def fib_recursive(n, steps):
    steps[0] += 1
    if n <= 1:
        return n
    return fib_recursive(n - 1, steps) + fib_recursive(n - 2, steps)

# Iterative Fibonacci function (for a single value)
def fib_iterative_single(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# Generate Fibonacci series using recursive approach
def fibonacci_series_recursive(n):
    series = []
    total_steps = [0]
    for i in range(n + 1):
        steps_for_i = [0]
        val = fib_recursive(i, steps_for_i)
        series.append(val)
        total_steps[0] += steps_for_i[0]
    return series, total_steps[0]

# Generate Fibonacci series using iterative approach
def fibonacci_series_iterative(n):
    if n < 0:
        return [], 0
    series = []
    steps = 0
    if n >= 0:
        series.append(0)
        steps += 1
    if n >= 1:
        series.append(1)
        steps += 1
    a, b = 0, 1
    for i in range(2, n + 1):
        steps += 1
        next_val = a + b
        series.append(next_val)
        a, b = b, next_val
    return series, steps

# Driver Code
if __name__ == "__main__":
    n = int(input("Enter the value of n: "))

    if n < 0:
        print("Please enter a non-negative integer.")
    else:
        # Recursive Series
        print("\n--- Recursive Approach ---")
        series_rec, steps_rec = fibonacci_series_recursive(n)
        print(f"Fibonacci series (recursive): {series_rec}")
        print(f"Total steps (recursive): {steps_rec}")

        # Iterative Series
        print("\n--- Iterative Approach ---")
        series_iter, steps_iter = fibonacci_series_iterative(n)
        print(f"Fibonacci series (iterative): {series_iter}")
        print(f"Total steps (iterative): {steps_iter}")
```
